There are many VNC software such as TeamViewer, Remmina, TigerVNC, UltraVNC, AnyDesk, and so on. Because RealVNC server is a default software of Raspberry Pi Desktop OS, we will use VNC Viewer of RealVNC to access Raspberry Pi from your laptop. Download and install VNC Viewer on your laptop.
